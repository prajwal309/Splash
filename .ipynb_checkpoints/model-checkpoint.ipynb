```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
val = np.random.normal(1, 0.0015, 1000)
t = np.linspace(0, 10, 1000)
plt.figure(figsize=(10,8))
plt.scatter(t, val)
```

```python
x = np.linspace(-1, 0.95, 195)
coeffs = np.array([1, 1.941, -0.175, -2.246, -1.125])
y = np.array([np.sum(np.array([coeffs[i]*(j**i) for i in range(len(coeffs))])) for j in x]) * 0.0075 
plt.plot(x, y)
plt.show()
```

```python
val = np.random.normal(1, 0.0015, 1000)
val[400:595] += y
t = np.linspace(0, 10, 1000)
plt.figure(figsize=(10,8))
plt.scatter(t, val)
```

```python
f_std = np.std(val)
f_med = np.median(val)
outlier_f = []
outlier_t = []
for t, f in zip(t, val):
    if f > f_med + 2.5 * f_std:
        outlier_f.append(f)
        outlier_t.append(t)
```
